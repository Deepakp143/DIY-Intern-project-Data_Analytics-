```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("C:\\Users\\praja\\Documents\\Projects\\DIY project\\1719219834-DoctorVisits - DA (1).csv")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['visits'], bins=30, kde=True)
plt.title('Distribution of Doctor Visits')
plt.xlabel('Number of Visits')
plt.ylabel('Frequency')
plt.show()
```

```python

# Gender distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='gender', data=data)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Income distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['income'], bins=30, kde=True)
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```

```python
# Relationship between visits and age
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='visits', data=data)
plt.title('Doctor Visits vs Age')
plt.xlabel('Age')
plt.ylabel('Number of Visits')
plt.show()
```

```python
# Relationship between visits and income
plt.figure(figsize=(10, 6))
sns.scatterplot(x='income', y='visits', data=data)
plt.title('Doctor Visits vs Income')
plt.xlabel('Income')
plt.ylabel('Number of Visits')
plt.show()
```

```python

# Health status distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='health', data=data)
plt.title('Health Status Distribution')
plt.xlabel('Health Status')
plt.ylabel('Count')
plt.show()
```

```python
# Relationship between health status and doctor visits
plt.figure(figsize=(10, 6))
sns.boxplot(x='health', y='visits', data=data)
plt.title('Doctor Visits by Health Status')
plt.xlabel('Health Status')
plt.ylabel('Number of Visits')
plt.show()
```

```python
# Pairplot
sns.pairplot(data)
plt.show()
```

```python
# Calculate the percentage of illness based on gender
illness_gender = data.groupby('gender')['illness'].value_counts(normalize=True).unstack() * 100
```

```python
# Print the percentage table
print(illness_gender)
```

```python
gender_illness_counts = data.groupby(['gender', 'illness']).size().unstack().fillna(0)
gender_illness_percentages = gender_illness_counts.div(gender_illness_counts.sum(axis=1), axis=0) * 100
```

```python
# Plot pie charts for illness distribution within each gender
for gender in gender_illness_percentages.index:
    plt.figure(figsize=(8, 6))
    plt.pie(gender_illness_percentages.loc[gender], labels=gender_illness_percentages.columns, autopct='%1.1f%%', startangle=140)
    plt.title(f'Percentage of Illness by Gender: {gender}')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()
```

```python
average_visits_by_age = data.groupby('age')['visits'].mean().reset_index()
```

```python
# Plot the average number of doctor visits by age
plt.figure(figsize=(10, 6))
sns.lineplot(data=average_visits_by_age, x='age', y='visits', marker='o')
plt.title('Average Number of Doctor Visits by Age')
plt.xlabel('Age')
plt.ylabel('Average Number of Visits')
plt.grid(True)
plt.show()
```

```python
# Visualizing Visits Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['visits'], bins=20, kde=False, color='blue')
plt.title('Distribution of Doctor Visits')
plt.xlabel('Number of Visits')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python

# Visualizing Visits by Gender
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='gender', order=data['gender'].value_counts().index, palette='viridis')
plt.title('Number of Visits by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Visits')
plt.show()
```

```python
# Age Distribution of Patients
plt.figure(figsize=(12, 6))
sns.histplot(data['age'], bins=20, kde=True, color='green')
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Private vs Free/Reduced Healthcare
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='private', hue='reduced', palette='Set1')
plt.title('Distribution of Private vs Free/Reduced Healthcare')
plt.xlabel('Private Healthcare')
plt.ylabel('Number of Visits')
plt.legend(title='Reduced')
plt.grid(True)
plt.show()
```

```python
# Analysis of Illness Types
plt.figure(figsize=(12, 8))
illness_counts = data['illness'].value_counts()
sns.barplot(x=illness_counts.index, y=illness_counts.values, palette='plasma')
plt.title('Distribution of Illness Types')
plt.xlabel('Illness')
plt.ylabel('Number of Visits')
plt.xticks(rotation=45)
plt.show()
```
